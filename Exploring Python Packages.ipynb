#Perform basic mathematical operations (e.g., element-wise addition, multiplication) on arrays.

```python
print("a + 5 =",a1+5)
print("a - 5 =",a1-5)
print("a * 5 =",a1*5)
print("a / 5 =",a1/5)
print("a % 5 =",a1%5)
print("a ** 5 =",a1**5)
print("a // 5 =",a1//5)
```

The code performs various arithmetic operations on the NumPy array a1 (which contains integers from 1 to 10) with the scalar value 5. Here's a breakdown of each operation:

a1 + 5: Adds 5 to each element of the array.

a1 - 5: Subtracts 5 from each element of the array.

a1 * 5: Multiplies each element of the array by 5.

a1 / 5: Divides each element of the array by 5 (resulting in floating-point values).

a1 % 5: Computes the remainder when each element is divided by 5.

a1 ** 5: Raises each element to the power of 5.

a1 // 5: Performs integer division, dividing each element by 5 and discarding the remainder.

#Calculate the mean, median, and standard deviation of a dataset.

```python
df = pd.read_csv("pima-indians diabetes.csv")
glucose=np.array(df["Glucose"])
print("Mean of Glucose =",glucose.mean())
print("Median of Glucose=",np.median(glucose))
print("Standard Deviation of Glucose =",glucose.std())
```

The code reads the "pima-indians diabetes.csv" dataset into a pandas DataFrame df. It then extracts the "Glucose" column from the DataFrame as a NumPy array glucose. The following operations are performed:

glucose.mean(): Computes the mean (average) of the 'Glucose' values.

np.median(glucose): Computes the median of the 'Glucose' values.

glucose.std(): Computes the standard deviation of the 'Glucose' values.

#Create a simple line plot of the function y = x^2 for x in the range of -10 to 10.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-10, 10, 100)
y = x**2
plt.plot(x, y, label='y = x^2')
plt.title('Plot of y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

The code uses matplotlib to plot the graph of y=x^2

**Imports:**

matplotlib.pyplot is imported for plotting the graph.

numpy is imported for creating a range of values for
x using np.linspace.

**x values:**

x = np.linspace(-10, 10, 100) creates 100 evenly spaced values for
x between -10 and 10.

**y values:**

y = x**2 computes the square of each value in x, giving the corresponding
y values for the equation y=x^2

**Plotting:**

plt.plot(x, y, label='y = x^2') plots the curve
y=x^2, with a label for the legend.

plt.title('Plot of y = x^2') sets the title of the plot.

plt.xlabel('x') and plt.ylabel('y') label the x and y axes.

plt.grid(True) adds a grid to the plot.

plt.legend() displays the legend with the label 'y = x^2'.

**Displaying the plot:**

plt.show() displays the plot.

The output is a graphical plot of the function
y=x^2, where the x-axis ranges from -10 to 10, and the y-axis shows the corresponding squared values.
